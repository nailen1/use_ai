## 4. 프롬프트 전송 테스트

```python
from use_ai import prompt_to_model

response = prompt_to_model(
    prompt="Python의 장점 3가지를 간단히 알려줘.",
    temperature=0.5,
)
print(response)
```

```python
response = prompt_to_model(
    prompt="오늘의 인사말을 한 줄로 해줘.",
    system_message="너는 투자관련 전문 AI 비서야.",
    max_tokens=100,
)
print(response)
```
